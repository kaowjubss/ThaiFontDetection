```python
import sys
import os
from pathlib import Path 
import cv2
import numpy as np


def rotation(img, start, stop, step):
  rows, cols = img.shape
  matrix = []
  for i in range(start, stop, step):
    i=i/10
    M = cv2.getRotationMatrix2D((cols/2,rows/2), i, 1)
    matrix.append(cv2.warpAffine(img, M, (cols, rows)))
  return matrix

def shearing(img, start, stop, step):
  rows, cols = img.shape
  matrix = []
  for i in range(start, stop, step):
    i=i/10
    M = np.float32([[1, 0, 0],
                    [np.tan(i * np.pi/180), 1, -(rows/1.5)*np.sin(i * np.pi/180)]])
    matrix.append(cv2.warpAffine(img, M, (cols, rows)))
  return matrix

def blurring(img, stop):
  rows, cols = img.shape
  matrix = []
  for i in range(1, stop, 2):
    ksize = i
    ker = np.ones((ksize,ksize),np.uint8)/(ksize*ksize)
    matrix.append(cv2.filter2D(img,0,ker))
  return matrix

def croped2square(roi,increase_ratio=1.2):
    h, w = roi.shape
    r=int(max(h,w)*increase_ratio)
    square = np.zeros((r,r), dtype=np.uint8)
    # background.fill(255)  # or img[:] = 255
    # place the image in the center of the background
    offset_x = int((square.shape[1] - w) / 2)
    offset_y = int((square.shape[0] - h) / 2)
    square[offset_y:offset_y+h, offset_x:offset_x+w] = roi
    return square

def cropfit(img):
    # find the bounding box of the image
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
    #find the first row, first col, last row, last col of white pixels
    x, y, w, h = cv2.boundingRect(img)
    #crop the image
    img = img[y:y+h, x:x+w]
    return img

```
